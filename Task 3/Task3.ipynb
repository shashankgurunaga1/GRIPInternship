As a business manager to find out weak areas where you can make more profit

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
url='SampleSuperstore.csv'
df=pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
#Checking the empty values in the dataframe

df.isnull().sum()
```

```python
#Checking the profit margin

df['Profit_Margin']=(df.Profit/df.Sales)*100
df.head()
```

```python
#Grouping by Sub-Category
p1=df.groupby(['Sub-Category'])
p1.first()
```

```python
#Grouping on the basis of the sub category and simultaneously fetching only the sales and profit values

pd1=pd.DataFrame(df.groupby(['Sub-Category'])[['Sales', 'Profit','Profit_Margin']].sum())
pd1
```

```python
sub_df= pd.DataFrame(pd1.sort_values('Profit_Margin', ascending = False))
sub_df
```

```python
#Dataframe consisting of data sorted on the basis of profit in descending order

subcat_profit_df2 = pd.DataFrame(sub_df.sort_values('Profit', ascending = False))

subcat_profit_df2
```

```python
#Dataframe consisting of data sorted on the basis of sales in descending order

subcat_sales_df2 = pd.DataFrame(sub_df.sort_values('Sales', ascending = False))

subcat_sales_df2
```

```python
sns.set_theme(style="whitegrid")


figure, axis = plt.subplots(1, 3, figsize=(12, 6))

subcat1 = sns.barplot(data = sub_df, x = sub_df.index, y = sub_df.Profit_Margin, ax=axis[0], palette = "bright")
subcat1.set(title="Most Profitable Sub-Category")
subcat1.set_xticklabels(subcat1.get_xticklabels(),rotation = "vertical", size = 10)




subcat2=sns.barplot(data=subcat_profit_df2,x=subcat_profit_df2.index,y=subcat_profit_df2.Profit,ax=axis[1],palette="bright")
subcat2.set(title="Best Selling Sub-Category")
subcat2.set_xticklabels(subcat2.get_xticklabels(),rotation = "vertical", size = 10)


subcat3=sns.barplot(data=subcat_sales_df2,x=subcat_sales_df2.index,y=subcat_sales_df2.Sales,ax=axis[2],palette="bright")
subcat3.set(title="Highest Selling Sub-Category")
subcat3.set_xticklabels(subcat3.get_xticklabels(),rotation = "vertical", size = 10)


figure.tight_layout()
plt.show()




```
